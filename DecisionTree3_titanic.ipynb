```python
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
inputs=df.drop('Survived',axis='columns')
target=df.Survived
```

```python
inputs.head()
```

```python
target
```

```python
inputs.Sex=inputs.Sex.map({'male': 1, 'female':2})
```

```python
inputs.Age[:10]
```

```python
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
```

```python
inputs.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(inputs,target,test_size=0.25)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn import tree
model= tree.DecisionTreeClassifier(criterion='entropy')
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
print('Wrong Prediction out of Total:')
print((y_test!=y_pred).sum(),'/',((y_test==y_pred).sum()))
```

```python
from sklearn.metrics import accuracy_score
print('Percentage Accuracy:',100*accuracy_score(y_test,y_pred))
```

```python
# confusion Matrix
from sklearn.metrics import confusion_matrix
```

```python
c_m=confusion_matrix(y_test,y_pred)
```
