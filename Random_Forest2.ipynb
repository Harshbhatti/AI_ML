```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
inputs=df.drop('Survived',axis='columns')
target=df.Survived
```

```python
inputs.head()
```

```python
target
```

```python
inputs.Sex=inputs.Sex.map({'male':1,'female':2})
```

```python
inputs.Age[:10]
```

```python
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
```

```python
inputs
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(inputs,target,test_size=0.25)
```

```python
len(x_test)
```

```python
len(x_test)
```

```python
sc=StandardScaler()
x_train_std=sc.fit_transform(x_train)
x_test_std=sc.fit_transform(x_test)
```

```python
x_train_std
```

```python
x_test_std
```

```python
forest=RandomForestClassifier(criterion='entropy',n_estimators=100,random_state=10)
```

```python
forest.fit(x_train_std,y_train)
```

```python
y_pred=forest.predict(x_test_std)
```

```python
y_pred
```

```python
print('wong predictions out of total:')
print((y_test!=y_pred).sum(),'/',((y_test==y_pred).sum()))
```

```python
print('Percentage Accuracy:',100*accuracy_score(y_test,y_pred))
```
