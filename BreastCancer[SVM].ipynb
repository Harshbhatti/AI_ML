# Import Python Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

# Let's view the data in a dataframe

```python
df_cancer=pd.DataFrame(np.c_[cancer['data'],cancer['target']],columns=np.append(cancer['feature_names'],['target']))
```

# Support Vactore Machine

# Train Model

```python
x=df_cancer.drop(['target'],axis=1)
x.head()
```

```python
y=df_cancer['target']
y.head()
```

# Create the training and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print('The size of our training"x"(input features)is',x_train.shape)
print('The size of our training"x"(input features)is',x_test.shape)
print('The size of our training"y"(input features)is',y_train.shape)
print('The size of our training"y"(input features)is',y_test.shape)
```

# Import Support Vector Machine(SVM) Model

```python
from sklearn.svm import SVC
```

```python
svc_model=SVC()
```

# Now,let's train our SVM model with our 'training' dataset.

```python
svc_model.fit(x_train,y_train)
```

# Let's use our trained model to make a prediction using our testing data

```python
y_predict=svc_model.predict(x_test)
```

# Confusion Metric

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=np.array(confusion_matrix(y_test,y_predict,labels=[1,0]))
confusion=pd.DataFrame(cm,index=['is_cancer','is_health'],columns=['peredicted_cancer','predicted_healthy'])
```

```python
confusion
```

```python
sb.heatmap(confusion,annot=True)
plt.show()
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predict)
```

```python
print('wrong Prediction out of Total:')
print((y_test!=y_predict).sum(),'/',((y_test==y_predict).sum()+y_test!=y_predict).sum()+(y_test!=y_predict).sum())
```

# Feature Scaling Training & Testing Data

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_train_scaled=sc.fit_transform(x_train)
```

```python
x_test_scaled=sc.fit_transform(x_test)
```

```python
svc_model=SVC()
svc_model.fit(x_train_scaled,y_train)
```

```python
y_predict=svc_model.predict(x_test_scaled)
cm=confusion_matrix(y_test,y_predict)
```
